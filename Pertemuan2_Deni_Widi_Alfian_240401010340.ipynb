```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

**Q2. Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?**

```python
missing = df.isnull().sum()
percent = (missing / len(df)) * 100

pd.DataFrame({
    'Missing': missing,
    'Percent (%)': percent
})
```

Jawaban:

Berikut adalah kolom yang memiliki missing value:

1. Kolom Cabin, sebanyak 687 baris kosong (77.1%)
2. Kolom Age,sebanyak 177 baris kosong (19.9%)
3. Kolom Embarked, sebanyak 2 baris kosong (0.2%)

**Q3. Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy
atau Pandas.**

```python
# Menggunakan NumPy
survived_arr = df['Survived'].to_numpy()
pct_selamat = np.mean(survived_arr) * 100
print(f"Selamat    : {np.sum(survived_arr)} orang")
print(f"Tidak      : {np.sum(survived_arr == 0)} orang")
print(f"% Selamat  : {pct_selamat:.2f}%")

# Verifikasi dengan Pandas
print(f"Pandas check: {df['Survived'].mean()*100:.2f}%")
```

Jawaban:

Dari dataset tersebut, hanya 38.38% dari 891 penumpang yang selamat.

---




**Q4. Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?**

```python
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

jumlah = len(wanita_kelas1)
rata_usia = wanita_kelas1['Age'].mean()
min_usia = wanita_kelas1['Age'].min()
max_usia = wanita_kelas1['Age'].max()

print(f"Jumlah penumpang wanita kelas 1        : {jumlah} penumpang")
print(f"Rata-rata usia penumpang wanita kelas 1: {rata_usia:.1f} tahun")
```

Jawaban:

Dari dataset tersebut, penumpang wanita dari kelas 1 sebanyak 94 penumpang, dengan rata-rata usia 34.6 tahun (~35 tahun).

**Q5. Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?**

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean()
survival_by_class
```

Jawaban:

Dari dataset tersebut, kelas 1 memiliki tingkat keselamatan tertinggi (62.96%), hampir 2.5× lebih tinggi dari Kelas 3 (24.24%).

**Q6. Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.**

```python
# Analisis multi-variabel: Sex × Pclass
pivot = df.groupby(['Pclass', 'Sex'])['Survived']\
          .mean().unstack().round(3) * 100
print(pivot)
```

Jawaban:

Dari dataset tersebut, wanita di semua kelas jauh lebih selamat daripada pria. Pria kelas 2 (15.7%) bahkan lebih rendah dari wanita kelas 3 (50%).
